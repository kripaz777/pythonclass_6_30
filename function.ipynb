```python
def input_value():
    l = int(input("Enter l = "))
    b = int(input("Enter b = "))
    return l,b

def area():
    ar = input_value()
    l,b = ar
    a = l*b
    return a

def output():
    print(area())
    

output()
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
    
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + or - or * or /  ")
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("Invalid operator")

cal()
```
